```python
import matplotlib.pyplot as plt
import random as rd
from CHAOS import Chaos

# Creation of dir where data is goingo to be stored
! mkdir data

# Parameters
N = 101# Number of stars
dt = 0.1 # Time step in Gyr
rc = 10 # Reference length

# Asymetry parameters
b = 1
c = 1

# Initial velocities
vx0 = 0

vy0_1 = 3.8
vy0_2 = 4.2
vy0 = rd.uniform(vy0_1,vy0_2)

vz0_1 = 3.8
vz0_2 = 4.2
vz0 = rd.uniform(vz0_1,vz0_2)

# Initial point
x0_1 = 2.5
x0_2 = 2.6
x0 = rd.uniform(x0_1,x0_2)
y0, z0 = 0, 0


plt.figure(figsize =(17,7))
plt
for k in range (0, N ):
    datos = Chaos(vx0, vy0, vz0, dt, b, c, rc)
    pos, vel, energy, poin, ace = datos.verlet(x0,y0,z0, k)
    plt.subplot(1,2,1)
    plt.xlabel('x(kPc)')
    plt.ylabel('v($km/s$)')
    plt.title('Poincare section for b = {} and c = {}'.format(b,c))
    plt.plot(poin[0], poin[1], 'b*', markersize=2)
    plt.subplot(1,2,2)
    plt.xlabel('t(Gyr)')
    plt.ylabel('E($km^2$/$s^2$)')
    plt.title('Energy of the orbits')
    plt.plot(energy[1], energy[0], '*', label = 'energy of particule '+str(k) , markersize=2)
    plt.savefig('data/Poincareb='+str(b)+' c='+str(c)+' N='+str(N)+' lim x0 = ['+str(x0_1)+','+str(x0_2)+'] lim vy0 = ['+str(vy0_1)+','+str(vy0_2)+'] lim vz0 = ['+str(vz0_1)+','+str(vz0_2)+' rc = '+str(rc)+'.png')
    #plt.legend()
    #plt.savefig('data/Energy b='+str(b)+' c='+str(c)+' N='+str(N)+'.png')
    

```
