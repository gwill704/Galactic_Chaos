```python
import matplotlib.pyplot as plt
import random as rd
import os
import numpy as np
from Chaos import Chaos

# Creation of dir where data is goingo to be stored
os.makedirs('data', exist_ok=True)

# Parameters
N = 100# Number of stars
dt = 0.1 # Time step in Gyr
rc = 20 # Reference length

# Asymetry parameters
b = 0.5
c = 0.2

# Energy

E = -61.

plt.figure(figsize =(17,7))

for k in range (0, N ):
    # input point in the three first arguments of chaos
    x0 = rd.uniform(0.2*rc/np.sqrt(np.e),0.5*rc/np.sqrt(np.e))
    omega = rd.uniform(np.sqrt(-2*E)/x0, 2*np.sqrt(-2*E)/x0)
    datos = Chaos(x0,0,0, dt, b, c, rc, E, omega)
    pos, vel, energy, poin= datos.verlet(k)
    plt.subplot(1,2,1)
    plt.xlabel('x(kPc)')
    plt.ylabel('v($km/s$)')
    plt.title('Poincare section for b = {} and c = {}'.format(b,c))
    plt.plot(poin[0], poin[1], 'b*', markersize=2)
    plt.subplot(1,2,2)
    plt.xlabel('t(Gyr)')
    plt.ylabel('E($km^2$/$s^2$)')
    plt.title('Energy of the orbits')
    plt.plot(energy[1], energy[0], '*', label = 'energy of particule '+str(k) , markersize=2)
    plt.savefig('data/Poincareb='+str(b)+' c='+str(c)+' N='+str(N) + '.png')
```

```python
# Plot of the orbit of the last star saved in the loop

plt.figure(figsize = (17,7))
plt.subplot(1,2,2)
plt.plot(pos[1], pos[2])
plt.xlabel('y (kpc)')
plt.ylabel('z (kpc)')
plt.title('Orbit of a star y-z, b = {}, c = {}'.format(b,c))
plt.subplot(1,2,1)
plt.plot(pos[0],pos[1])
plt.xlabel('x (kpc)')
plt.ylabel('y (kpc)')
plt.title('Orbit of a star x-y, b = {}, c = {}'.format(b,c))
plt.savefig('data/Orbitb='+str(b)+' c='+str(c)+' N='+str(N)+'.png')#+' lim x0 = ['+str(x0)+'] lim E0 = ['+str(E)+' Lz = '+str(Lz)+ 'rc = '+str(rc)+'.png')
```
